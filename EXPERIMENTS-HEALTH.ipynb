# #-------------------------------------------------------------------------------------------------

## Data Preprocessing

```python
diabetes = pd.read_csv('data/diabetes.csv')

diabetes
```

```python
# normalization 0-1
for i in diabetes.columns:
    diabetes[i] = (diabetes[i] - diabetes[i].min()) / (diabetes[i].max() - diabetes[i].min())
```

```python
# split in test and train data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes.drop('Outcome',axis=1), diabetes['Outcome'])
```

```python
X_train = X_train.values.tolist()
X_test = X_test.values.tolist()
y_train = y_train.values.tolist()
y_test = y_test.values.tolist()
```

```python
y_train = [int(x) for x in y_train]
y_test =  [int(x) for x in y_test]
```

# #----------------------------------------------------------------------------

### Train and Test with Bias

```python
weightVectorsEncoding, weightVectorsHSGS, weightVectorsClassico, weightVectorsClassicoBin = experiment_TRAIN(X_train, 
                                                                                                             y_train,
                                                                                                             lrParameter=0.2, 
                                                                                                             n_epochs=200, 
                                                                                                             trainingBias=True,
                                                                                                             trainingApproaches={'neuronio-classico', 'neuronio-classico-bin', 'hsgs','encoding-weight'})
```

```python
resultsBIAS = experiment_TEST(Xs_test, 
                          ys_test, 
                          weightVectorsEncoding, 
                          weightVectorsHSGS, 
                          weightVectorsClassico, 
                          weightVectorsClassicoBin, 
                          repeat=30,
                          bias=True, 
                          testingApproaches={'neuronio-classico', 'neuronio-classico-bin', 'hsgs','encoding-weight'})
```

```python
with open('testesout/outputs/HEALTH_ZERO.json', 'w') as fp:
    json.dump(resultsBIAS, fp)
```
